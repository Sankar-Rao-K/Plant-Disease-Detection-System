```python
import os
import hashlib
from PIL import Image

def find_and_remove_duplicates(folder_path):
    hashes = {}
    duplicates = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                file_path = os.path.join(root, file)
                try:
                    # Open and hash image
                    with Image.open(file_path) as img:
                        img_hash = hashlib.md5(img.tobytes()).hexdigest()

                    if img_hash in hashes:
                        print(f"Duplicate found and removed: {file_path}")
                        os.remove(file_path)
                        duplicates += 1
                    else:
                        hashes[img_hash] = file_path
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")
    print(f"\n Done. Removed {duplicates} duplicates from {folder_path}\n")

# Run for each split
find_and_remove_duplicates("Data_sets/Train")
find_and_remove_duplicates("Data_sets/validate")
find_and_remove_duplicates("Data_sets/test")
```
